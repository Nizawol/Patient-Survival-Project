### Introduction  
Cancer is one of the top causes of death globally. Survival of cancer patients has been a big challenge for many years. In this project, a dataset of liver cancer patients will be analyzed to predict the survival rate. 

### About the Data 

The data contains 155 observations and 20 features of hepatitis patients. The features are AGE, SEX, STEROID, ANTIVIRALS, FATIGUE, MALAISE, ANOREXIA, LIVER_BIG, LIVER_FIRM, SPLEEN_PALPABLE, SPIDERS, ASCITES, VARICES, BILIRUBIN, ALK_PHOSPHATE, SGOT, ALBUMIN, PROTIME, HISTOLOGY and Class. 6 of these features are numerical and the rest are chategorical.

### Data transformation


#### Skewness 

```python
cancer_data_updated.skew()
```

some of the features like 'ALK_PHOSPHATE', 'BILIRUBIN' and 'SGOT' are skewed to the right. We have to transform them inorder to get reliable result. Other features such as 'AGE', 'ALBUMIN' and 'PROTIME' are not skewed significantly, so it can be ignored. There are a number of methods to transform here log transformation will be used.  

```python
features_transform = ['ALK_PHOSPHATE', 'BILIRUBIN', 'SGOT']

cancer_data_updated[features_transform] = cancer_data_updated[features_transform].applymap(np.log)

cancer_data_updated.skew()
```

As can be seen from skewness output, we can see that the skewness present in our data was mainly fixed.

#### Transforming categorical data to numerical - Dummy variables

```python
# let us drop age_group column scince it has the same information with age
cancer_data_cleared = cancer_data_cleared.drop('AGE_GROUP', axis = 1)

cancer_data_cleared.head()
```

```python
# Creating dummy variables for the categorical features

dummy = cancer_data_cleared.select_dtypes(exclude=[np.number]).columns


cancer_data_dummy = pd.get_dummies(cancer_data_cleared, prefix = dummy, columns = dummy, drop_first = True)
cancer_data_dummy.head()
```

### Building Models

```python
# Splitting the dataset into the Training set and Test set

X = cancer_data_dummy.drop('Class_LIVE', axis = 1)
y = cancer_data_dummy.Class_LIVE
    
from sklearn.model_selection import train_test_split

training_features, test_features, training_target, test_target, = \
train_test_split(X,y, test_size = .2, random_state= 45)
```

```python
# let us check the partition
print(training_features.shape)
print(test_features.shape)
print(training_target.shape)
```

#### Using a Dummy Classifier
As a first classifier, it good to apply the built-in DummyClassifier class from sklearn.dummy to set a baseline for performance of other models.

```python
from sklearn.dummy import DummyClassifier

dummy_classifier = DummyClassifier(strategy="most_frequent")

dummy_classifier.fit(test_features, test_target)
```

```python
prediction_score = dummy_classifier.score(test_features, test_target)
print('The prediction score is: {:5.3f} %'.format(prediction_score*100))
```

#### Decission Tree machine learning 

```python
# importing the model, fitting to training features and its target variables

from sklearn.tree import DecisionTreeClassifier
DT_Model = DecisionTreeClassifier(criterion = 'entropy', random_state = 45)

DT_Model.fit(training_features, training_target)  
```

```python
# Prediction 
DT_prediction = DT_Model.predict(test_features) 

DT_prediction
```

```python
# Confusion Matrix - Comparing the predicted target value to the orginal target variable

# from sklearn.model_selection import cross_val_score, cross_val_predict
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
DT_confusion_m = confusion_matrix(test_target, DT_prediction)
print (DT_confusion_m)
```

```python
# Accuracy Score
DT_accuracy = accuracy_score(test_target, DT_prediction)

print(" Decision Tree Prediction Accuracy : {:.2f}%".format(DT_accuracy * 100))
```

#### Ensemble Methods - Random Forest

```python
# importing the model and Fitting to the Training set

from sklearn.ensemble import RandomForestClassifier

RF_Model = RandomForestClassifier( n_estimators=80, 
                               max_features=7, 
                               oob_score=True, 
                               random_state= 45)
RF_Model.fit(training_features, training_target)


# Predicting the Test set results
RF_prediction = RF_Model.predict(test_features)
```
